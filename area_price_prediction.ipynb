```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from  sklearn.metrics import r2_score ,mean_squared_error,mean_absolute_error
import pickle
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("area_price.csv")
```

```python
data.head()
```

```python
X=data[["area"]]
y=data["price"]
x=data["area"]
```

```python
plt.xlabel("area in Sqft",color="red",fontweight ="bold",fontsize="20")
plt.ylabel("price in INR ",color="green",fontweight ="bold",fontsize="20")
plt.title("relation between area and price",color="blue",fontweight ="bold",fontsize="20")
plt.plot(X,y,marker="o")
plt.show
```

```python
plt.xlabel("area in Sqft",color="red",fontweight ="bold",fontsize="20")
plt.ylabel("price in INR ",color="green",fontweight ="bold",fontsize="20")
plt.title("relation between area and price",color="blue",fontweight ="bold",fontsize="20")
plt.bar(x,y)
plt.show()

```

```python

plt.scatter(X,y)
plt.show()
```

```python
model=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=45)
model.fit(x_train,y_train)
```

```python
prredicted_value=model.predict(x_test)
```

```python
r2=r2_score(y_test,prredicted_value)
```
